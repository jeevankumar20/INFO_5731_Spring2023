```python
import re
import nltk
from nltk.corpus import stopwords

def clean_tweet(text):
    # Remove non-ASCII characters
    text = text.encode('ascii', 'ignore').decode()

    # Remove URLs
    text = re.sub(r'http\S+', '', text)

    # Remove special characters, symbols, and digits
    text = re.sub(r'[^\w\s]|[\d]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text

new['Text'] = new['Text'].apply(clean_tweet)

# Removing Stopwords
stop = stopwords.words('english')
new['Text'] = new['Text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))

# After cleaning
new['word_count'] = new['Text'].apply(lambda x: len(str(x).split(" ")))

# Number of stop words
nltk.download('stopwords')
stop = stopwords.words('english')
new['stopwords'] = new['Text'].apply(lambda x: len([x for x in x.split() if x in stop]))

# Number of special Characters
def count_special_chars(text):
    special_chars = string.punctuation  # include all !"#$%&\'()*+,-./:;<=>?@[\\]^_{|}~ punctuation characters and space
    return sum([1 for char in text if char in special_chars])
new['Special Characters'] = new['Text'].apply(count_special_chars)

# Number of numerics
new['numerics'] = new['Text'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))

new1 = new
new.head()
```
