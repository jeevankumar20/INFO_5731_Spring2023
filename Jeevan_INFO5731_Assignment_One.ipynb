# Question 1

(20 points). Write a Python program to generate 12 random numbers between 1 and 100. Sort the list of numbers in ascending order and print it out. Sort the list of numbers in descending order and print it out. Calculate the average for the group and print it out.

```python
# Your code here
import random
num=[]
for i in range(12):
    num.append(random.randint(1,100))
print(num)
num.sort()
print("Ascending Order:",num)
num.sort(reverse=True)
print("Descending Order:",num)
avg=sum(num)/len(num)
print("Average:",avg)
```

# Question 2
(25 points). Write a program that will do the following string formatting tasks:

(1) Ask the user to enter two sentences.

(2) Make each sentence into a list. Each element in the list should correspond to a word in the sentence. For example, if the sentence is "Alas! Am I alive? This is my first python program", the program should print out ['Alas, '!', 'Am', 'I', 'alive, '?', 'This', 'is', 'my', 'first', 'python', 'program'].

(3) Compare the two sentences. Print out a list of words that occur in both sentences.

(4) Remove any punctuation from the sentence and print it back out. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”)

(5) Count the number of vowels in the sentences. Print out each vowel and the number of times it appears in the sentences, such as: a:2, e:1, i:0, etc

```python
# Your code here
import string
import re


s1=input("Enter Sentence One: ")
s2=input("Enter Sentence Two: ")


#Hello, I am Jeevan Kumar!!!

#Alas! Am I alive? This is my first python program.



#Each element as a list
l1 = [word for word in re.findall(r'\b\w+\b|[^\w\s]', s1)]
l2 = [word for word in re.findall(r'\b\w+\b|[^\w\s]', s2)]
print("Sentence One into List: ",l1)
print("Sentence Two into List: ",l2)


#Getting only common words from both sentences
word_list1 = [word.lower() for word in re.findall(r'\b\w+\b', s1)]
word_list2 = [word.lower() for word in re.findall(r'\b\w+\b', s2)]
common_words = [word for word in word_list1 if word in word_list2]
print("Words that occur in both sentences:", common_words)


#Removed punctuatuion from sentences
print("Sentence1 without punctuation:"," ".join(word_list1))
print("Sentence2 without punctuation:"," ".join(word_list2))


#count each vowel in both sentences
def count_vowels(sentence):
    vowels = "aeiou"
    vowel_count = {v:0 for v in vowels}
    for char in sentence:
        if char in vowels:
            vowel_count[char] += 1

    for vowel, count in vowel_count.items():
        print(f"{vowel}: {count}")
print("Vowels in First Sentence: ")
count_vowels(s1.lower())
print("Vowels in Second Sentence: ")
count_vowels(s2.lower())
```

(15 points). The formula that describes the volume V of a sphere with radius r is the following:

$ V=\frac{4}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of 1–10. Output the result in the following format:

	r	V
	1	…
	2	…
	3	…


```python
# Your code here
import math

print("r\tV")
for r in range(1, 11):
    V = (4/3)*math.pi*(r**3)
    print("{}\t{:.2f}".format(r,V))

```

```python
import pandas as pd

df = pd.read_csv('/Assignment1_denton_housing.csv')
df
```

```python
#Getting only Occupied Housing Units
occupied=df.loc[df['title_field'] == 'Occupied Housing Units']
occupied
```

```python
#Finding the Difference between Occupied Housing Units from year to year.
occ_vals=occupied['value'].values
diffs_a=occ_vals[:-1]-occ_vals[1:]
years=["2014-2013","2013-2012","2012-2011","2011-2010","2010-2009","2009-2008"]
df2=pd.DataFrame()
df2['years']=years
df2['Difference']=abs(diffs_a)
print(df2)
#Sorted Difference 
print("Sorting the difference from year to year: ")
sorted_df = df2.sort_values(by='Difference')
print(sorted_df)
```

```python
print(occupied)
print(occ_vals)
vacant=df.loc[df['title_field']=='Vacant Housing Units']
vac_vals=vacant['value'].values
print(vacant)
print(vac_vals)
total=df.loc[df['title_field']=='Total Housing Units']
tot_vals=total['value'].values
print(total)
print(tot_vals)
perc_of_occupied=[]
perc_of_vacant=[]
for i in range(7):
    perc_of_occupied.append(round(((occ_vals[i]/tot_vals[i])*100),2))

for i in range(7):
    perc_of_vacant.append(round(((vac_vals[i]/tot_vals[i])*100),2))

year=occupied['year'].values
print("Percentage of Vacant and Occupied Housing units for all years: ")
print("Year\tVacant Housing Units\tOccupied Housing Units")
for i in range(7):
    print("{}\t\t{:.2f}\t\t{:.2f}".format(year[i],perc_of_vacant[i],perc_of_occupied[i]))
```

```python

maxvac=max(vac_vals)
maxocc=max(occ_vals)
maxyearocc=occupied[occupied['value']==maxocc]['year'].item()
maxyearvac=vacant[vacant['value']==maxvac]['year'].item()
print("Highest number of housing units that were vacant and occupied:")
print("\t\t\tYear\t\tValue")
print("Vacant Housing Units:\t{}\t\t{}".format(maxyearvac,maxvac))
print("Occupied Housing Units:\t{}\t\t{}".format(maxyearocc,maxocc))
```

```python
print("Total Housing Units:", end="\n")
print(total)
print('Values:',tot_vals)
reciprocals = [1/x for x in tot_vals]
harmonic_mean = len(tot_vals) / sum(reciprocals)
print("Harmonic Mean: ",harmonic_mean)
```
